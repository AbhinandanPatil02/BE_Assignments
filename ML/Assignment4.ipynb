```python
def func(x):
    return (x+3)**2

def der_x(x):
    return 2*(x+3)
```

```python
gd=[]
def gradient_decent(start_x,learning_rate,iterations):
    x=start_x
    gd.append(x)
    for i in range(iterations):
        slope=der_x(x)
        x=x-learning_rate*slope
        gd.append(x)
        print("Iterations : ",i+1,"   X : ",x,"  f(x) : ",func(x))
    return x
```

```python
start_x=3
l_r=0.1
itr=50
ans= gradient_decent(3,l_r,50)
print("Local minima occurs at : ",ans)
```
